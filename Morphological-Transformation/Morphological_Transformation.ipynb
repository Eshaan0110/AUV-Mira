```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


```

## 1. Erosion

**Erosion** removes pixels from the boundaries of objects.

**How it works:**
- The kernel moves across the image.
- A pixel remains white (foreground) **only if all pixels under the kernel are white**.
- If any pixel under the kernel is black, the pixel becomes black.

**Effect:**
- Shrinks white regions
- Removes small white noise
- Breaks thin connections between objects


```python
img = cv2.imread('ten.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
kernel = np.ones((20,20),np.uint8)
kernel
```

```python
#Erosion

erosion= cv2.erode(img,kernel,iterations=1)
plt.imshow(erosion)
```

## 2. Dilation

Dilation adds pixels to the boundaries of objects.

### How it works

- The kernel moves across the image.
- A pixel becomes white (foreground) if **at least one pixel under the kernel is white**.
- If all pixels under the kernel are black, the pixel remains black.

### Effect

- Expands white regions
- Fills small holes inside objects
- Connects nearby or broken objects

```python
#Dilation

dilation = cv2.dilate(img,kernel,iterations=1)
plt.imshow(dilation)
```

## 3. Opening

Opening is a combination of **erosion followed by dilation**.

### How it works

- First, **erosion** removes small white pixels and shrinks objects.
- Then **dilation** expands the remaining objects back to near their original size.

### Effect

- Removes small white noise
- Separates objects that are slightly connected
- Smooths object boundaries while preserving the overall shape

```python
#morphologyEX

opening = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.imshow(opening)
```

## 4. Closing

Closing is a combination of **dilation followed by erosion**.

### How it works

- First, **dilation** expands objects and fills small gaps.
- Then **erosion** shrinks the objects back to their approximate original size.

### Effect

- Fills small holes inside objects
- Closes small gaps between objects
- Smooths object boundaries

```python
#Closing/ MORPH_CLOSE

closing = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
plt.imshow(closing)
```

## 5. Morphological Gradient

Morphological Gradient finds the **difference between dilation and erosion**.

### How it works

- The image is first **dilated** and then **eroded**.
- The eroded image is subtracted from the dilated image.
- This highlights the boundary pixels of objects.

### Effect

- Detects edges of objects
- Highlights object outlines
- Shows the boundaries between foreground and background

```python
#Morphology gradient

gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
plt.imshow(gradient)
```
